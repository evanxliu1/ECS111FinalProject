4 rotations × 4 model variants = 16 runs.

Each run takes ~30–45 min on a T4. The full sweep is ~10 hours; split across multiple Colab sessions, checkpoints persist to Drive.

```python
import os
from pathlib import Path

CONFIGS = ['base', 'contrastive', 'adversarial', 'both']
ROTATIONS = [0, 1, 2, 3]

for config in CONFIGS:
    for rot in ROTATIONS:
        ckpt = Path(f'checkpoints/{config}/rotation_{rot}/best.pt')
        if ckpt.exists():
            print(f'skip (already done): {config} rotation {rot}')
            continue
        print(f'\n>>> training {config} rotation {rot}')
        ret = os.system(f'python -m src.train --config configs/{config}.yaml --rotation {rot}')
        if ret != 0:
            print(f'FAILED: {config} rotation {rot}, stopping early')
            break
```
